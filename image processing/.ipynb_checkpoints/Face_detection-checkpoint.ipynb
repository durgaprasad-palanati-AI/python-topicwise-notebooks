```python
import cv2
cascPath = "haarcascade_frontalface_default.xml"
```

```python
# Create the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)
# Read the image
image = cv2.imread('faces.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30)
    #flags = cv2.CV_HAAR_SCALE_IMAGE
)

print("Found {0} faces!".format(len(faces)))
```

```python
import matplotlib.pyplot as plt
```

```python
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

#plt.imshow("Faces found", image)
imgplot = plt.imshow(image)
#plt.show()
plt.savefig('faces_rect.jpg',dpi=800)
```

# EYE detection

```python
eyecascpath='C:/Users/durga/AppData/Local/Programs/Python/Python37/Lib/site-packages/cv2/data/haarcascade_eye.xml'
# Create the haar cascade
eyeCascade = cv2.CascadeClassifier(eyecascpath)
# Read the image
image = cv2.imread('durga.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detect eyes in the image
eyes = eyeCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5
    #flags = cv2.CV_HAAR_SCALE_IMAGE
)

print("Found {0} eyes!".format(len(eyes)))
```

```python
#Draw a rectangle around the eyes
for (x, y, w, h) in eyes:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(image)
#plt.show()
plt.savefig('durga_eyes_rect.jpg',dpi=800)
```
