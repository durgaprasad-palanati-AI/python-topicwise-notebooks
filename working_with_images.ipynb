# using OPENcv

```python
import cv2
from  matplotlib import pyplot as plt
```

# Method-1

```python
im = cv2.imread('index.jpg')
im.shape    # height:  275
            #width:   183
            # channel: 3
```

```python
im_resized = cv2.resize(im, (224, 224), interpolation=cv2.INTER_LINEAR)

plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

# Method-2

```python
img = cv2.imread('index.jpg')		# this is read in BGR format
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)		# this converts it into RGB

plt.imshow(rgb_img)
plt.show()
```

# resize images
 Interpolation Method for Resizing  

    cv2.INTER_AREA: This is used when we need to shrink an image.
    cv2.INTER_CUBIC: This is slow but more efficient.
    cv2.INTER_LINEAR: This is primarily used when zooming is required. 
    This is the default interpolation technique in OpenCV.

```python
img = cv2.imread('index.jpg',1) # Loading the image
```

# reduce to half

```python
#reduce to half
half = cv2.resize(img,(0,0),fx = 0.5, fy = 0.5) 
plt.imshow(cv2.cvtColor(half,cv2.COLOR_BGR2RGB))
plt.show()
```

# En-large

```python
enlarge = cv2.resize(img, (1500, 1600)) 
plt.imshow(cv2.cvtColor(enlarge,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
enlarge.shape
```

# Strech

```python
stretch = cv2.resize(img, (780, 540),
               interpolation = cv2.INTER_NEAREST)
plt.imshow(cv2.cvtColor(stretch,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
stretch_cubic = cv2.resize(img, (780, 540),
               interpolation = cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(stretch_cubic,cv2.COLOR_BGR2RGB))
plt.show()
```

# Gray scale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```
